# Relatório de Análise VI

```python
import pandas as pd
dados = pd.read_csv('dados/aluguel_residencial.csv',sep=';')
dados.head()
```

### Vamos criar uma nova variável chamada de 'Valor bruto' que será a soma dos gastos (Valor +IPTU+ CONDOMINIO)

```python
dados['Valor Bruto'] = dados['IPTU'] + dados['Condominio'] + dados['Valor']
dados.head()
```

### Criação de uma variável que diga o valor do m² (Valor/area)

```python
dados['Valor m2'] = dados['Valor']/dados['Area']
dados.head()
```

### Ajustando o grau de precisão

```python
dados['Valor m2'] = dados['Valor m2'].round(2)
dados.head()
```

### Criando uma coluna 'Valor Bruto do m2' (Valor Bruto/area)

```python
dados['Valor Bruto m2'] = (dados['Valor Bruto']/dados['Area']).round(2)
dados.head()
```

### Criaremos uma variável que abrigue os tipos de imóvel casa e apartamento, para tanto, devemos coletar a variável Tipo e recolher esses marcadores e indentifica-los.

```python
casa = ['Casa', 'Casa de Condomínio', 'Casa de Vila']
dados['Tipo Agregado'] = dados['Tipo'].apply(lambda x: 'Casa' if x in casa else 'Apartamento')
dados.head()
```

## Excluindo Variáveis

```python
dados_aux = pd.DataFrame(dados[['Tipo Agregado', 'Valor m2', 'Valor Bruto', 'Valor Bruto m2']])
```

```python
dados_aux.head()
```

## Temos varias formas de eliminar as colunas, algumas opções:

```python
del dados_aux['Valor Bruto']
dados_aux.head()
```

```python
dados_aux.pop('Valor Bruto m2')
dados_aux.head()
```

```python
dados.drop(['Valor Bruto', 'Valor Bruto m2'], axis = 1, inplace = True)
dados.head()
```

```python
dados.to_csv('dados/aluguel_residencial.csv',sep=';',index= False)
```
